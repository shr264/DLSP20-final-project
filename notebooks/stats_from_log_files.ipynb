```python
def convertLogFilestoCsv():
    import glob, os
    import pandas as pd
    log_files = list(glob.glob("*.log"))
    print(list(log_files))
    for log_file in log_files:
        try:
            print('Processing {}'.format(log_file))

            train_loss = []
            loss = []
            truep = []
            falsep = []
            falsen = []

            with open(log_file) as fp:
                line = fp.readline()
                cnt = 1
                new_cnt = 1
                while line:
                    text = line.strip()
                    if 'Epoch' in text and 'Average loss:' in text:
                        print("Line {} : {}".format(cnt, text))
                        train_loss.append(float(text.split('Average loss: ')[1]))
                    if 'Test set: Average loss' in text:
                        print("Line {} : {}".format(cnt, text))
                        loss.append(float(text.split('Test set: Average loss: ')[1].split(',')[0]))
                    if 'Class 1' in text:
                        print("Line {} : {}".format(cnt, text))
                        line = fp.readline()
                        cnt += 1
                        text = line.strip()
                        print("Line {} : {}".format(cnt, text))
                        if 'TP' in line and 'FP' in line and 'FN' in line:
                            print('TP:{}'.format(float(text.split('TP')[1].split(',')[0])))
                            truep.append(float(text.split('TP')[1].split(',')[0]))
                            print('FP:{}'.format(float(text.split('FP')[1].split(',')[0])))
                            falsep.append(float(text.split('FP')[1].split(',')[0]))
                            print('FN:{}'.format(float(text.split('FN')[1])))
                            falsen.append(float(text.split('FN')[1]))
                    cnt += 1
                    line = fp.readline()
                print("Line {} : {}".format(cnt, text))
                print(new_cnt)
                
                n = min(len(train_loss), len(loss), len(truep), len(falsep), len(falsen))

                pd.DataFrame({log_file.split('.log')[0]+'train_loss':train_loss[:n],
                    log_file.split('.log')[0]+'loss':loss[:n],
                log_file.split('.log')[0]+'tp':truep[:n],
                log_file.split('.log')[0]+'fp':falsep[:n],
                log_file.split('.log')[0]+'fn':falsen[:n]}).to_csv(
                    log_file.split('.log')[0]+'_stats.csv')
        except Exception as e: 
            print(e)
```

```python
convertLogFilestoCsv()
```
